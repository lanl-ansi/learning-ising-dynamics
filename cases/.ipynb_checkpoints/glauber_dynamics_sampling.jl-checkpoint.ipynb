```julia
function count_freq(A)
    d = Dict{Array{eltype(A),1}, Int}()
    for i in 1:size(A,1)
        a = A[i,:]
        if a in keys(d)
            d[a] += 1
        else
            d[a] = 1
        end 
    end
    return d
end
```

```julia
d = count_freq(samples_pairs_iid)
```

```julia
spin_samples = Array{Int64,2}(undef,length(d),8)
a = [key for key in keys(d)]
for i = 1:length(a)
    vect = a[i]
    spin_samples[i,:] = vcat(d[vect],vect[1],2*digits(vect[2],base=2,pad=spin_number) .- 1,2*digits(vect[3],base=2,pad=spin_number) .- 1)'
end
spin_samples
```

```julia
pairs_configs = []
spin_number = 3
config_number = 2^spin_number

for i = 0:(config_number - 1)
    for j = 0:(config_number - 1)
        sigma_i = 2*digits(i, base=2, pad=spin_number).-1
        sigma_j = 2*digits(j, base=2, pad=spin_number).-1
        temp_ij = sigma_i.*sigma_j
        if length(findall(isequal(-1),temp_ij)) <= 1
            push!(pairs_configs,[i,j])
        end
    end
end
display(pairs_configs)
```

```julia
function dictionary_samples_hist(samples_T1::Array)
    d1 = Dict{Array{eltype(samples_T1),1}, Int}()
    for i in 1:size(samples_T1,1)
        a = samples_T1[i,2:end]
        if a in keys(d1)
            d1[a] += samples_T1[i,1]
        else
            d1[a] = samples_T1[i,1]
        end
    end
    return d1
end
```

```julia
d1 = dictionary_samples_hist(samples_T1)
```

```julia
d2 = dictionary_samples_hist(samples_T2)
```

```julia
for a in keys(d2)
    if a in keys(d1)
        d1[a] += d2[a]
    else
        d1[a] = d2[a]
    end
end
```
